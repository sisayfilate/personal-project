### Factorial

```python
def factorial(n):
    f = 1
    for i in range(n):
        f = f * (n-i)
    return f
factorial(4)
```

### GroupBy

```python
import pandas as pd
raw_data = {'employee_name': ['Andy', 'Beth', 'Cindy', 'Dale'], 
        'employee_id': [123456, 789456, 654123, 963852], 
        'date_joined': ['2015-02-15', '2014-02-15', '2017-05-16', '2018-01-15'], 
        'age': [45, 36, 34, 25],
        'yrs_of_experience' :  [24, 15, 14, 4]}
df = pd.DataFrame(raw_data, columns = ['employee_name', 'employee_id', 'date_joined', 'age', 'yrs_of_experience'])
df
```

```python
df['experience_binned'] = df['yrs_of_experience'].apply(lambda x: "low" if x < 6 else ("Medium" if x<16 else "High"))
df
```

```python
df['yrs_remain'] = df['age'].apply(lambda x: 65-x)
df
```

```python
df.loc[df['employee_name']=='Cindy', ['employee_name']]='Cris'
df
```
